```python
#Exercice 7 (Si l'image n'est pas en négatif, relancer le programme, il affiche d'abord l'image original puis pour la deuxième exècution l'image en négative)
from PIL import Image,ImageDraw
from IPython.display import display
#taille image
largeur =639
hauteur=480

img=Image.open("iguane.jpg")
for x in range (largeur):
    for y in range (hauteur):
        r,v,b=img.getpixel((x,y))
        nr,nv,nb=255-r,255-v,255-b
        img.putpixel((x,y),(nr,nv,nb))
        #print(x,y,nr,nv,nb)
img.save("iguane.jpg")
display(img)
```
